```python
import random
import numpy as np
```

```python
with open('amazon_list.txt','r') as f:
    amazon=f.readlines()
with open('dslr_list.txt','r') as f:
    dslr=f.readlines()
with open('webcam_list.txt','r') as f:
    webcam=f.readlines()
```

```python
a_tr=random.sample(amazon,int(len(amazon)*0.9))
a_ts=list(set(amazon)-set(a_tr))
with open('a_tr.txt','a+') as f:
    for i in a_tr:
        f.write(i)
with open('a_ts.txt','a+') as f:
    for i in a_ts:
        f.write(i)
```

```python
d_tr=random.sample(dslr,int(len(dslr)*0.9))
d_ts=list(set(dslr)-set(d_tr))
with open('d_tr.txt','a+') as f:
    for i in d_tr:
        f.write(i)
with open('d_ts.txt','a+') as f:
    for i in d_ts:
        f.write(i)
```

```python
w_tr=random.sample(webcam,int(len(webcam)*0.9))
w_ts=list(set(webcam)-set(w_tr))
with open('w_tr.txt','a+') as f:
    for i in w_tr:
        f.write(i)
with open('w_ts.txt','a+') as f:
    for i in w_ts:
        f.write(i)
```

```python
with open('a_tr.txt','r') as f:
    atr=f.readlines()

with open('d_tr.txt','r') as f:
    dtr=f.readlines()
```

```python
a_fish=random.sample(atr,int(len(atr)*0.1))
with open('a_fish.txt','a+') as f:
    for i in a_fish:
        f.write(i)
```

```python
d_fish=random.sample(dtr,128)
with open('d_fish.txt','w') as f:
    for i in d_fish:
        f.write(i)
```
